```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")#dataset is 
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split

# Create a LightGBM Regressor with the correct objective for regression
regressor = lgb.LGBMRegressor(objective='regression', n_estimators=50)  # You can adjust the number of estimators (n_estimators) as needed

# Fit the LightGBM Regressor to the training data
regressor = regressor.fit(X_train, y_train)

# Now, you can use the trained LightGBM Regressor to make predictions
y_pred = regressor.predict(X_test)
```
